### Data cleaning and preparation

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings 
warnings.simplefilter("ignore")
```

```python
#reading the csv file  aka uploading
df = pd.read_csv("Cardiotocographic.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
num_cols_with_miss_values = ['LB','AC','DS','DP','MLTV','Width','Tendency','NSP']
for columns in num_cols_with_miss_values:
    mean_value = df[columns].mean()
    df[columns].fillna(mean_value,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
num_columns = df.select_dtypes(include=['float64','int64']).columns
num_columns
```

```python
for column in num_columns:
    Q1=df[column].quantile(0.25)
    Q3=df[column].quantile(0.75)
    IQR=Q3-Q1
    l_bound=Q1-1.5*IQR
    u_bound=Q3+1.5*IQR
    outliers=df[(df[column]<l_bound)|(df[column]>u_bound)]
    if len(outliers):
        print(f'The Outliers present in column {column} are :')
        print(outliers[[column]])
    
```

```python
df=df[(df[column]>=l_bound)&(df[column]<=u_bound)]
df
```

```python
df.reset_index(inplace= True)
df
```

```python
df.drop('index',axis=1,inplace=True)
df
```
