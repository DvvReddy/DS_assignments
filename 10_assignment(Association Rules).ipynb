# Association Rules

### Data Preprocessing

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
import warnings 
warnings.simplefilter("ignore")
```

```python
df=pd.read_csv("Online retail.csv")
df
```

```python
df.dropna()
df.drop_duplicates(inplace=True)
df.tail(15)
```

```python
df.columns=["group of things"]
splitting=df["group of things"].str.split(",").apply(lambda x:[groups.strip()for groups in x if groups])
encoder= TransactionEncoder()
df_encoder= encoder.fit(splitting).transform(splitting)
TE_data=pd.DataFrame(df_encoder,columns=encoder.columns_)
df.tail(15)
```

```python
TE_data.tail(15)
```

### Association Rule Mining

```python
frequent_things= apriori(TE_data,min_support=0.01,use_colnames=True)
frequent_things.tail(15)
```

```python
num_itemsets = 5 

ass_rules = association_rules(frequent_things, 
                        metric="confidence", 
                        min_threshold=0.2, 
                        num_itemsets=num_itemsets) 

ass_rules = ass_rules[ass_rules['lift']>1.0]
ass_rules
```

```python
ass_rules_sorting=ass_rules.sort_values(by='lift',ascending=False)
ass_rules_sorting[['antecedents', 'consequents', 'support', 'confidence', 'lift']].tail()
```

### INterview Questions

#### What is Lift and how does it help analyze associations?

Lift is a metric that  helps us to assess the strength and significance of relationships between items in our data. It compares the observed co-occurrence of items (A and B) to what would be expected by random chance.

Lift > 1: Indicates a positive association, meaning A and B often appear together more frequently than expected by chance.
Lift < 1: Suggests a negative association, where A and B rarely co-occur.
Lift = 1: Implies no significant relationship between the items.

#### What is Support and how does it identify frequent itemsets?

Support is a  metric that reflects how common an itemset (combination of items) is within our data. It's calculated as the proportion of transactions containing the entire itemset relative to the total number of transactions.

High Support: Indicates a frequently occurring itemset, suggesting a potential association between the items.
Low Support: Suggests a less common combination, which might not be as interesting for analysis.

#### What is Confidence and how does it relate to prediction?

Confidence is a metric which  focuses on the predictive power of an association rule. It tells us how likely we'll find item B in a transaction if we already know item A is present.

High Confidence: Suggests a strong correlation, meaning finding A often leads to finding B as well. This can be helpful for making predictions.
Challenges of Association Rule Learning

Scalability: As datasets grow larger, the number of potential itemsets can explode, leading to significant computational demands for processing all the combinations.

Threshold Selection: Choosing the right support and confidence thresholds is crucial. Setting them too low can generate an overwhelming number of potentially irrelevant rules, while setting them too high might miss valuable insights.

Relevancy Filtering: Not all discovered rules are equally interesting or actionable. Identifying and filtering out trivial or irrelevant rules can be a challenge.
