### Talking about the scenario 

Standardization is preferred when:
#### Data is normally distributed or the algorithm assumes a Gaussian distribution.   
Many machine learning algorithms, such as linear regression, logistic regression, and support vector machines, perform better when the data is standardized. Â  
#### Outliers are present. 
Standardization is less sensitive to outliers than normalization, as it doesn't bound the data to a specific range. 


Normalization (Min-Max Scaling) is preferred when:
#### Data needs to be bounded within a specific range, such as between 0 and 1. 
This is often useful for neural networks, where input values are expected to be within a certain range. Â  
#### The distribution of the data is unknown or not Gaussian.
Normalization doesn't make any assumptions about the underlying data distribution. Â  
#### Distance-based algorithms are used. 
Normalization can be beneficial for algorithms like k-nearest neighbors (KNN), where distances between data points are important. Â

```python
encoder =OneHotEncoder()
dummy=pd.get_dummies(df[["sex",'income']],dtype=int)
dummy
```

```python
labels =LabelEncoder()
```

```python
df['education'] = labels.fit_transform(df['education'])
df['workclass'] = labels.fit_transform(df['workclass'])
df['marital_status'] = labels.fit_transform(df['marital_status'])
df['occupation'] = labels.fit_transform(df['occupation'])
df['relationship'] = labels.fit_transform(df['relationship'])
df['race'] = labels.fit_transform(df['race'])
df['native_country'] = labels.fit_transform(df['native_country'])
```

```python
df.drop(df[['sex','income']],axis=1,inplace=True)
df=df.join(dummy,how='right')
df
```

### Feature Engineering

```python
df['capital-gain-minus-loss'] = df['capital_gain'] - df['capital_loss']
df['age_squared'] = df['age'] ** 2
df
```

### Feature Selection of the model

```python
isolation_forest=IsolationForest(random_state=15,contamination=0.1)
isolation_forest.fit(df[num_cols])
```

```python
presence_of_outliers =isolation_forest.predict(df[num_cols])
```

```python
cleaned_df = df[presence_of_outliers != -1]
```

```python
print("shape of Original dataset :", df.shape)
print("shape of Cleaned dataset:", cleaned_df.shape)
```
