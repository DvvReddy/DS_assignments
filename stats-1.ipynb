## standardization of numerical variables

##### Standardization, also known as z-score normalization, is a technique used to transform data so that it has a mean of 0 and a standard deviation of 1. This process adjusts the values of the data based on how far they deviate from the mean, measured in units of standard deviation. The formula for standardization is:(x – μ) / σ where:   ( x ) is the original value, ( \mu ) is the mean of the data, ( \sigma ) is the standard deviation of the data, ( z ) is the standardized value.

```python
num_columns = sales_discount_data.select_dtypes(exclude=['object']).columns
num_columns
```

```python
plt.figure(figsize=(15,8),edgecolor= 'grey')
plt.title('after using standardization')
for columns in num_columns:
    sns.histplot(standardization[columns],label=columns)
plt.legend()
plt.show()
```

### conversion of categorical to dummy variables 

Many machine learning algorithms, like those used in neural networks or linear regression, work best with numbers. However, real-world data often includes categorical data, such as colors, countries, or product types.

To make this categorical data usable for these algorithms, we transform it into a numerical format called one-hot encoding. This technique creates a new binary column for each category. If a data point belongs to a specific category, the corresponding column is set to 1; otherwise, it's 0.

By using one-hot encoding, we can prevent the algorithm from mistakenly assuming an order or hierarchy among categories. This helps the model learn the true relationships between features and the target variable, leading to better predictions and insights
